```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("AviationData.csv", encoding='latin1', low_memory=False)
```

```python
df['Event.Date'] = pd.to_datetime(df['Event.Date'], errors='coerce')
```

```python
df['Year'] = df['Event.Date'].dt.year
```

```python
df = df[df['Year'] >= 1983]
```

```python
df['Year'].min()
```

```python
df[['Total.Fatal.Injuries', 
    'Total.Serious.Injuries', 
    'Total.Minor.Injuries', 
    'Total.Uninjured']].head()
```

```python
cols = ['Total.Fatal.Injuries', 
        'Total.Serious.Injuries', 
        'Total.Minor.Injuries', 
        'Total.Uninjured']

for col in cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df[cols] = df[cols].fillna(0)
```

```python
df['Total.Injuries'] = (
    df['Total.Fatal.Injuries'] +
    df['Total.Serious.Injuries'] +
    df['Total.Minor.Injuries']
)
```

```python
df[['Total.Fatal.Injuries', 
    'Total.Serious.Injuries', 
    'Total.Minor.Injuries', 
    'Total.Injuries']].head()
```

```python
make_counts = df['Make'].value_counts()

valid_makes = make_counts[make_counts >= 50].index

df_filtered = df[df['Make'].isin(valid_makes)]
```

## Key Findings

1. Large aircraft have higher average injuries than small aircraft due to carrying more passengers.

2. Weather conditions significantly affect accident severity:
   - IMC (poor weather) leads to more injuries than VMC.

3. Phase of flight impacts injury outcomes:
   - Climb and approach phases have higher injuries.
   - Landing and taxi phases are less severe.

4. Aircraft manufacturer differences exist, but filtering was necessary to remove rare categories.
